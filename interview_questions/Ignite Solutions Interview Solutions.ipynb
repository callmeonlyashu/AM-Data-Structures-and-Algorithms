```python
"""
Most languages have a built in sort method that will sort an array of strings alphabetically.
Demonstrate how to sort an array of strings by the length of each string, shortest strings first.

"""

# - Create a dictionary with string and its length.
# - Sort the dictionary by its string length
# - Get list of string back from dictionary

def sortString( lst = [ 'This', 'is', 'the', 'demo', "string" ] ):
    
    dictWithLength = dict( zip( lst, list( map( lambda x: len(x), lst ) ) ) )
    
    return list( {k: v for k, v in sorted(dictWithLength.items(), key=lambda item: item[1])}.keys() )
```

```python
sortString()
```

```python
"""
Write a simple program that reads a line from the keyboard and outputs the same line where
every word is reversed. A word is defined as a continuous sequence of alphanumeric characters
or hyphen (‘-’). For instance, if the input is

“We are at Ignite Solutions! Their email-id is careers@ignitesol.com”
the output should be
“eW era ta etingI snoituloS! riehT di-liame si sreerac@losetingi.moc”
HINT: ​Read the problem and the example carefully before starting.

"""

def reverseStringWithoutSpecialCharacters(x):
    
    rev = list(x)
    l = 0
    r = len(rev) - 1

    while l < r:
        if( not rev[l].isalpha() ):
            l += 1
        elif( not rev[r].isalpha() ):
            r -= 1
        else:
            rev[l], rev[r] = rev[r],  rev[l]
            l += 1
            r -= 1

    return '' . join( rev )

def reverseString():
    rev = 'We are at Ignite Solutions! Their email-id is careers@ignitesol.com'
    
    print( ' '. join( list( map( reverseStringWithoutSpecialCharacters, rev.split(' ') ) ) ) )
```

```python
reverseString()
```

```python
"""
Consider the following pattern:

A → D M → P X → A
a → d m → p x → a

Now, write a program to solve the following message using Python, JavaScript or Ruby.
Vrphwklqj phdqlqjixo
Hint: The answer is something meaningful.

"""

""" 
It seems something wrong with the pattern.  
For something meaningful it should be like D-A, P-M etc.

"""

def getEncodeCharacter(x):
    
    if x.isalpha() and (ord(x) - 3) < 97:
        return chr( ord(x) - 3 + 26 )
    elif x.isalpha() and ord(x) != 32:
        return chr( ord(x) - 3 )
    elif not x.isalpha():
        return x
    else:
        return ' '

def encodeStringWithPattern( string = 'Vrphwklqj@ phdqlqjixo' ):
    
    
    return ''.join( list( map( getEncodeCharacter,  string.lower() ) ) )
    
```

```python
encodeStringWithPattern()
```

```python
"""
Program to sort a list of integers
"""

from random import randint
lst = [ randint(1, 100 ) for x in range( 1,1000000 ) ]

# Counting sort

def countingSort(array1, max_val):
    m = max_val + 1
    count = [0] * m                
    
    for a in array1:
    # count occurences
        count[a] += 1             
    i = 0
    for a in range(m):            
        for c in range(count[a]):  
            array1[i] = a
            i += 1
    return array1

countingSort( lst, max( lst ) )
```
