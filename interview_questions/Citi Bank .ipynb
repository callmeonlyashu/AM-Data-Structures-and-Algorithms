```python
def double_decorator(func):
    @wraps(func) 
    def my_func(*args, **kwargs):
        res = func(*args, **kwargs)
        return res*2
    return my_func


@double_decorator
def calculate_sum(*args, **kwargs):
    """
    Double the response from the function
    """
#     @wraps(func) 
    res = sum(args)
    res += sum(kwargs.values())
    return res

```

```python
v1 = calculate_sum(1,2,3)
print(calculate_sum.__doc__)
```
