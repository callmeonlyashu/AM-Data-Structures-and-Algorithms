```python
"""
	Input: [3, 4, -1, 1]
Output: 2
Why: The array has 1, 3, 4 — the smallest missing positive integer is 2.

⸻

	2.	Input: [1, 2, 0]
Output: 3
Why: 1 and 2 are present, so the next smallest missing positive is 3.

⸻

	3.	Input: [7, 8, 9, 11, 12]
Output: 1
Why: There’s no 1 in the array, so it’s the smallest missing positive.

⸻

	4.	Input: [-5, -3, -2]
Output: 1
Why: No positive integers at all, so start from 1.

⸻

	5.	Input: [2, 1]
Output: 3
Why: 1 and 2 are there, next is 3.

"""

def smallest_positive(num):
    if 1 not in num:
        return 1
    
    max_number = max(num) + 2
    for i in range(1, max_number):
        if i not in num:
            return i
        

def test_smallest_positive():
    assert smallest_positive([3, 4, -1, 1]) == 2
    assert smallest_positive([1, 2, 0]) == 3
    assert smallest_positive([7, 8, 9, 11, 12]) == 1
    assert smallest_positive([-5, -3, -2]) == 1
    assert smallest_positive([2, 1]) == 3
    assert smallest_positive([1, 1, 2, 2]) == 3
    assert smallest_positive([1]) == 2
    assert smallest_positive([]) == 1
    assert smallest_positive([5, 3, 2, 1]) == 4
    assert smallest_positive([1, 2, 3, 4, 5]) == 6
    print("All tests passed!")

test_smallest_positive()
```

```python
from collections import defaultdict

def count_subsequences_combinatorics(oldFile, newFile):
    # Step 1: Preprocess: store all indices for each character in oldFile
    char_indices = defaultdict(list)
    for i, ch in enumerate(oldFile):
        char_indices[ch].append(i)

    def dfs(pos, index, path):
        """
        pos   : current position in newFile (which character we are matching)
        index : minimum index in oldFile we can consider
        path  : list of selected indices (to show the sequence path)
        """
        if pos == len(newFile):
            print(f"✅ Match found: {path}")
            return 1  # Base case: we've matched the entire newFile

        ch = newFile[pos]
        total = 0
        if ch not in char_indices:
            return 0  # Character not present in oldFile at all

        for i in char_indices[ch]:
            if i >= index:
                print(f"Trying '{ch}' at index {i} for newFile[{pos}]")
                total += dfs(pos + 1, i + 1, path + [i])
        return total

    return dfs(0, 0, [])

oldFile = "ababc"
newFile = "abc"
result = count_subsequences_combinatorics(oldFile, newFile)
print(f"\nTotal ways to form '{newFile}' from '{oldFile}': {result}")
```
