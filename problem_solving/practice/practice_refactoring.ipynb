```python
def calc(d):
    r = {}
    for k in d:
        if k in r:
            r[k] += len(d[k])
        else:
            r[k] = len(d[k])
    s = 0
    for i in r:
        s += r[i]
    avg = s / len(r)
    m = 0
    for i in r:
        if r[i] > m:
            m = r[i]
            nm = i
    print('avg:', avg)
    print('max:', nm, m)
```

```python
data = {
    "apple": ["red", "green"],
    "banana": ["yellow"],
    "mango": ["yellow", "orange", "green"]
}

calc(data)
```

```python
# Refactoring: 

def calc(d:dict) -> None:
    """
    This method takes dictionary as a input and prints
    average of counts of the values and key and count of largest count 
    
    Params:
        d: dictionary with key and values as list
    
    Return:
        None
    """
    
    # Create a dict with key and count of its values
    dict_val_counts = {k:len(v) for k,v in d.items() }
    
    all_elements = sum(list(dict_val_counts.values()))
    avg = all_elements/len(dict_val_counts)
    
    # Sorted in reverse order to get the key with largest count
    max_element = sorted(dict_val_counts.items(), key=lambda x:x[1], reverse=True)
    max_key, max_count = max_element[0]

    print('avg:', avg)
    print('max:', max_key, max_count)
```

```python
data = {
    "apple": ["red", "green"],
    "banana": ["yellow"],
    "mango": ["yellow", "orange", "green"]
}

calc(data)
```
