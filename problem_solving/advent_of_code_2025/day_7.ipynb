## Part 2

### Copied... (Pascal's Triangle Approach)

```python
import sys
from collections import Counter

splits, paths = 0, Counter()
for line in grid:
    print("".join(line))
    for (i, c) in enumerate(line):
        match c:
            case 'S':
                paths[i] = 1
            case '^':
                if i in paths:
                    splits += 1
                    paths[i-1] += paths[i]
                    paths[i+1] += paths[i]
                    del paths[i]

print(splits, paths.total())
```

### Copied DFS Approach

```python
from functools import lru_cache

ROWS = len(grid)
COLS = len(grid[0])

# Find S
start_r, start_c = beam_path[0]

@lru_cache(None)
def dfs(r, c):
    # Out of bounds
    if c < 0 or c >= COLS:
        return 0
    
    # If went past bottom: one complete timeline
    if r >= ROWS:
        return 1

    cell = grid[r][c]

    # Empty space → go straight down
    if cell == '.' or cell == 'S':
        return dfs(r + 1, c)

    # Splitter → split timelines
    if cell == '^':
        left = dfs(r + 1, c - 1)
        right = dfs(r + 1, c + 1)
        return left + right

    return 0

# Final answer
print(dfs(start_r + 1, start_c))
```
