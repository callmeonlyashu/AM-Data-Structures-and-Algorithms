## Part 1

```python
total = 0
for b in banks:
    bat = list(b)
    bat = [int(i) for i in bat]
    largest = max(bat[:len(bat)-1])
    index_of_largest = bat.index(largest)
    second_largest = max(bat[index_of_largest+1:])
    the_num = str(largest) + str(second_largest)
    total += int(the_num)

print(total)
```

## Part 2

```python
# import itertools
total = 0

# banks = ['811111111111119']
# print(len(banks[0]))

loop_break = 0

res = []
for b in banks:
    bat = list(b)
    bat = [(idx, int(i)) for idx, i in enumerate(bat)]

    window_size = (len(bat)-12)+1
    og_window_size = (len(bat)-12)+1
    curr_max_index = 0
    taken_indices = []
    missing_indices = []


    all_nums = []
    while True:
        # Window size is 1 or 0 means, take all till end as nothing left to compare
        if window_size <= 1:
            tp = [p[1] for p in bat[curr_max_index:]]
            all_nums.extend(tp)
            all_nums = [str(p) for p in all_nums]
            # print(all_nums)
            new_num = ''.join(all_nums)
            res.append(int(new_num))
            break

        # If window reaches end, stop
        if (curr_max_index+window_size) > len(bat):
            all_nums = [str(p) for p in all_nums]
            new_num = ''.join(all_nums)
            res.append(int(new_num))
            break
        
        curr_window = bat[curr_max_index:curr_max_index+window_size]
    
        # Find max_element and max_index in current window
        curr_max = 0
        for e in curr_window:
            if curr_max < e[1]:
                curr_max = e[1]
                curr_max_index = e[0]

        all_nums.append(curr_max)

        # Register taken indices
        taken_indices.append(curr_max_index)
    
        # Resiger missing index till now
        k = 0
        while k <= curr_max_index:
            if k not in missing_indices and k not in taken_indices:
                missing_indices.append(k)
            k += 1


        # Shrink window
        window_size = og_window_size - len(missing_indices)

        curr_max_index = curr_max_index + 1

print(sum(res)) # 169347417057382
```
