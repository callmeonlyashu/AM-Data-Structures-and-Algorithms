## Day 1 - Part 1

```python
inp = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
"""



# with open("day_1_input_A.txt") as f:
#     inp = f.read()
```

```python
prev_position = "50"
prev_direction = "R"

inp_arr = inp.splitlines()
# print(inp_arr)

count_zeros = 0
for i in inp_arr:
    curr_direction, rotation = i[0], i[1:]
    # Rotation should be  in module of 100 as it resets after every 100
    if int(rotation) > 100:
        rotation = int(rotation) % 100

    # If rotation is L, substract HANDLING 100
    if curr_direction == "L":
        if int(rotation) > int(prev_position):
            curr_position = 100-(int(rotation) - int(prev_position))
        else:
            curr_position = int(prev_position) - int(rotation)
    else:
        # If rotation is R, add HANDLING 100
        curr_position = int(rotation) + int(prev_position)
        if curr_position >= 100:
            curr_position = curr_position - 100

    if curr_position == 0:
        count_zeros += 1

    prev_position, prev_direction = curr_position, curr_direction

print(f"COUNT ZEROS: {count_zeros}")
    
```

## Day 1 - Part 2

```python
inp = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
"""



with open("day_1_input_A.txt") as f:
    inp = f.read()
```

```python
curr_position = 50

inp_arr = inp.splitlines()

count_zeros = 0
for i in inp_arr:
    global curr_position
    
    curr_direction, rotation = i[0], int(i[1:])
    if curr_direction == "L":
        i = 1
        while i <= rotation:
            
            if curr_position == 0:
                curr_position = 100

            curr_position = int(curr_position) - 1
            if curr_position == 0:
                count_zeros += 1

            i += 1
    else:
        i = 1
        while i <= rotation:

            curr_position = int(curr_position) + 1
            if curr_position == 100:
                curr_position = 0

            if curr_position == 0:
                count_zeros += 1


            i += 1


print(f"COUNT ZEROS: {count_zeros}")
    
```
