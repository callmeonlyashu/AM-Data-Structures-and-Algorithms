```python
def get_all_neighbours(test_input, index):
    row, col = index
    total_rows = len(test_input)
    total_cols = len(test_input[0])
    curr_index_val = test_input[row][col]
    all_nodes = []

    col_num = 1
    for _ in range(3):
        row_num = 1
        for _ in range(3):
            curr_row, curr_col = (row-row_num), (col-col_num) 

            row_num -= 1
            if curr_row < 0:
                continue

            if curr_col < 0:
                continue
            
            if curr_row < total_rows and curr_col < total_cols and (curr_row, curr_col) != index:
                node = test_input[curr_row][curr_col]
                all_nodes.append(node)
                
        col_num -= 1
        
    return all_nodes
    
# get_all_neighbours(test_input, (0, 0))
```
