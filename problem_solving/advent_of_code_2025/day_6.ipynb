## Part 1

### Prepare the grid to make it easy

```python

rows = inp.split("\n")
grid = [r.split(" ") for r in rows]

new_grid = []
for i in range(len(grid)):
    new_grid.append([])
    for j in range(len(grid[i])):
        if grid[i][j] not in [""]:
            if grid[i][j] not in ["+", "*"]:
                new_grid[i].append(int(grid[i][j]))
            else:
                new_grid[i].append(grid[i][j])

# print(new_grid)
```

### Actual logic

```python
total_rows, total_columns = len(new_grid), len(new_grid[0])

result = 0
for i in range(total_columns):
    col_total = 1
    for j in range(total_rows-1, -1, -1):
        if new_grid[j][i] in ["+", "*"]:
            operation = new_grid[j][i]
            if operation == "+":
                col_total = 0
            continue

        if operation == "+":
            col_total += new_grid[j][i]
        elif operation == "*":
            col_total *= new_grid[j][i]

    result += col_total


print(result)
```
