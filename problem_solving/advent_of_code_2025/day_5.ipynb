```python
with open("day_5_input.txt") as f:
    inp = f.read()
```

```python
data = inp.split("\n")

ranges = []
for i in data:
    if i == '':
        break

    l, r = i.split("-")
    ranges.append((int(l),int(r)))

ranges = sorted(ranges, key=lambda x: x[0])
low, high = ranges[0]

new_range = []
for r in range(1, len(ranges)):
    curr_low, curr_high = ranges[r]
    if high >= curr_low:
        high = max(high, curr_low, curr_high)
    else:
        new_range.append([low, high])
        low, high = ranges[r]

new_range.append([low, high])
```

```python
IDs = []
for i in data:
    if i == '' or "-" in i:
        continue

    IDs.append(int(i))

```

## Day 1

```python
fresh = 0
for i in IDs:
    for r in ranges:
        if r[0] <= i <= r[1]:
            fresh += 1
            break

print(fresh)
```

## Day 2

```python
count = 0
for r in new_range:
    low, high = r[0], r[1]
    count += (high-low)+1

print(count)
```
