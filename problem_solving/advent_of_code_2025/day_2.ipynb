```python
all_ranges = inp.split(",")
print(all_ranges)
```

## Day 2

```python
numbers = []
for r in all_ranges:
    bounds = r.split("-")
    low, high = int(bounds[0]), int(bounds[1])
    # print("low, high", low, high)
    for num in range(low, high+1):
        str_num = str(num)

        i, j = 0, 0
        while j <= len(str_num):
            half_num = str_num[i:j+1]
            j += 1
            times = 2
            while int(half_num*times) <= int(str_num):
                # print(f"{half_num*times} == {str_num}")
                if half_num*times == str_num and int(str_num) not in numbers:
                    numbers.append(int(str_num))
                times += 1

# print(numbers)
print(sum(numbers))
```
