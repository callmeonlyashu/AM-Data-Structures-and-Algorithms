Link: https://leetcode.com/problems/buy-two-chocolates/

```python
from typing import List

class Solution:
    def buyChoco(self, prices: List[int], money: int) -> int:
        prices.sort()
        # If the number of chocolates are less than 2 means you can't buy two chocolates
        if len(prices) < 2:
            return 0

        minimum_prices = prices[0] + prices[1]
        saving = money - minimum_prices
        if saving >= 0:
            return saving
        else:
            return money
        
```
