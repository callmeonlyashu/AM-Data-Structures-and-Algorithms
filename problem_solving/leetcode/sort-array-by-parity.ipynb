Link: https://leetcode.com/problems/sort-array-by-parity/

```python
from typing import List
class Solution:
    def sortArrayByParity(self, nums: List[int]) -> List[int]:
        odds = []
        evens = []

        for n in nums:
            if n % 2 == 0:
                evens.append(n)
            else:
                odds.append(n)

        
        return evens + odds
```
