### Tabulation

```python
from typing import List
class Solution:
    def rob(self, nums: List[int]) -> int:
        if len(nums) <= 2:
            return max(nums)

        # For 0 to n-1
        nums1 = nums[:len(nums)-1]
        dp_1 = [0]*len(nums1)
        dp_1[0] = nums1[0]
        dp_1[1] = max(nums1[0], nums1[1])

        for i in range(2, len(nums1)):
            dp_1[i] = max(dp_1[i-1], nums1[i]+dp_1[i-2])

        # From 1 to n
        nums2 = nums[1:len(nums)]
        dp_2 = [0]*len(nums2)
        dp_2[0] = nums2[0]
        dp_2[1] = max(nums2[0], nums2[1])

        for i in range(2, len(nums2)):
            dp_2[i] = max(dp_2[i-1], nums2[i]+dp_2[i-2])
        
        return max(dp_1[-1], dp_2[-1])
```
