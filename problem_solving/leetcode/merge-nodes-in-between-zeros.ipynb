```python
from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeNodes(self, head: Optional[ListNode]) -> Optional[ListNode]:
        curr = head.next

        res = ListNode(0)
        new_head = res

        curr_sum = 0
        while curr:
            if curr.val == 0:
                new_head.next = ListNode(curr_sum)
                curr_sum = 0
                new_head = new_head.next
            else:
                curr_sum += curr.val
            
            curr = curr.next
        return res.next
```
