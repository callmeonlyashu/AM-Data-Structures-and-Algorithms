```python
from typing import List

class Solution:
    def permuteUnique(self, nums: List[int]) -> List[List[int]]:
        res = []

        def backtrack(cur_arr, tracker):
            if len(cur_arr) == len(nums):
                if cur_arr.copy() not in res:
                    res.append(cur_arr.copy())
                return

            for i in range(len(nums)):
                if not tracker[i]:
                    cur_arr.append(nums[i])
                    tracker[i] = True
                    backtrack(cur_arr, tracker)
                    cur_arr.pop()
                    tracker[i] = False

        backtrack([], [False]*len(nums))
        return res
```
