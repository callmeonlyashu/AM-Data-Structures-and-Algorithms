Link: https://leetcode.com/problems/minimum-string-length-after-removing-substrings/

```python
class Solution:
    def minLength(self, s: str) -> int:
        if "AB" in s:
            s = s.replace("AB", "")
            return self.minLength(s)

        if "CD" in s:
            s = s.replace("CD", "")
            return self.minLength(s)

        return len(s)
        
```
