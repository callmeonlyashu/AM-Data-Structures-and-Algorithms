```python
from typing import List
from collections import deque
class Solution:
    def floodFill(self, image: List[List[int]], sr: int, sc: int, color: int) -> List[List[int]]:
        m, n = len(image), len(image[0])

        # Starting pixel
        color_queue = deque([(sr, sc)])
        visited = [(sr, sc)]

        while color_queue:
            row, col = color_queue.popleft()
            prev_color = image[row][col]
            image[row][col] = color

            # append adjacents to the queue
            if row-1 >= 0:
                coord1 = (row-1, col)
                if coord1 not in visited and prev_color == image[coord1[0]][coord1[1]]:
                    color_queue.append(coord1)
                visited.append(coord1)
            
            if row+1 < m:
                coord2 = (row+1, col)
                if coord2 not in visited and prev_color == image[coord2[0]][coord2[1]]:
                    color_queue.append(coord2)
                visited.append(coord2)

            if col+1 < n:
                coord3 = (row, col+1)
                if coord3 not in visited and prev_color == image[coord3[0]][coord3[1]]:
                    color_queue.append(coord3)
                visited.append(coord3)
            
            if col-1 >= 0:
                coord4 = (row, col-1)
                if coord4 not in visited and prev_color == image[coord4[0]][coord4[1]]:
                    color_queue.append(coord4)
                visited.append(coord4)

            # print(color_queue)
        
        # print(image)
        # print(visited)
        return image
```
