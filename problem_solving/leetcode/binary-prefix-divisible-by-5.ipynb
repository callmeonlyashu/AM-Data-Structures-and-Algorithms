Link: https://leetcode.com/problems/binary-prefix-divisible-by-5/

```python
from typing import List
class Solution:
    def prefixesDivBy5(self, nums: List[int]) -> List[bool]:
        cur_num = ""
        res = []
        for i in range(len(nums)):
            cur_num += str(nums[i])
            if int(cur_num, 2) % 5 == 0:
                res.append(True)
            else:
                res.append(False)

        return res
```
