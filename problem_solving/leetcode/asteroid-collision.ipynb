```python
from typing import List
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        stack = []

        for ast in asteroids:
            stack.append(ast)

            # Check Collision Possiblities
            collision_possible = True
    
            while collision_possible and len(stack) >= 2:
                last, second_last = stack[-1], stack[-2]

                # Collision possible only when they are moving towards each other
                if (last < 0 and second_last > 0): 
                    # Remove both the elements
                    stack.pop()
                    stack.pop()
                    # Re-assign after collision
                    if abs(last) > abs(second_last):
                        stack.append(last)
                    elif abs(last) < abs(second_last):
                        stack.append(second_last)
                    else:
                        pass
                else:
                    collision_possible = False

        return stack
```
