Link: https://leetcode.com/problems/count-operations-to-obtain-zero/

```python
class Solution:
    def countOperations(self, num1: int, num2: int) -> int:
        def rec(n1, n2, count):
            if n1 == 0 or n2 == 0:
                return count

            if n1 >= n2:
                n1, n2 = n1-n2, n2
            else:
                n1, n2 = n1, n2-n1

            count = rec(n1,n2, count+1)
            return count

        count = rec(num1, num2, 0)
        return count
```
