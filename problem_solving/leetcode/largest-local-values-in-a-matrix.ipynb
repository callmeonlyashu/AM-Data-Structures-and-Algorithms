```python
from typing import List
class Solution:
    def largestLocal(self, grid: List[List[int]]) -> List[List[int]]:
        n = len(grid)
        i, j = 0, 3
        k, l = 0, 3

        res = [[0] * (n - 2) for _ in range(n - 2)]
        p, q = 0, 0

        step_size = 1 
        while l <= n:
            i, j = 0, 3
            while j <= n:
                curr_grid = [g[i:j] for g in grid[k:l]]
                flat_list = list(itertools.chain.from_iterable(curr_grid))
                res[p][q] = max(flat_list)
                i += step_size
                j += step_size
                q += 1
            k += step_size
            l += step_size
            q = 0
            p += 1

        return res
```
