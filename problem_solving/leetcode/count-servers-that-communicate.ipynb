Link: https://leetcode.com/problems/count-servers-that-communicate/

```python
class Solution:
    def countServers(self, grid: List[List[int]]) -> int:
        servers = []
        for row in range(len(grid)):
            for col in range(len(grid[row])):
                if grid[row][col] == 1:
                    servers.append((row, col))

        # print(f"servers: {servers}")
        
        servers_set = set()
        for i in range(len(servers)):
            for j in range(i+1, len(servers)):
                row1, col1 = servers[i]
                row2, col2 = servers[j]
                # print(f"row1, col1", row1, col1 )
                # print("row2, col2", row2, col2)
                if (row1 == row2) or (col1 == col2):
                    servers_set.add((row1, col1))
                    servers_set.add((row2, col2))

        return len(servers_set)

```
