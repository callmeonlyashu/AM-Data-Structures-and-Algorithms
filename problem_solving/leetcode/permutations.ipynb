```python
from typing import List

class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        res = []

        def backtracking(curr_arr, track=[]):
            if len(curr_arr) == len(nums):
                res.append(curr_arr.copy())
                return

            for i in range(len(nums)):
                if not track[i]:
                    curr_arr.append(nums[i])
                    track[i] = True
                    backtracking(curr_arr, track)
                    curr_arr.pop()
                    track[i] = False
        
        backtracking([], [False]*len(nums))
        return res

```
