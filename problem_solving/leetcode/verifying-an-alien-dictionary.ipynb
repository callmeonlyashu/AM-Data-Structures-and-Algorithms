```python
class Solution:
    def isAlienSorted(self, words: List[str], order: str) -> bool:
        words_length = [len(w) for w in words]
        max_loop_length = max(words_length)

        order = " " + order
        def check_lex(chars):
            indexes = []
            for ch in chars:
                indexes.append(order.index(ch))
            
            sorted_index = sorted(indexes)
            # Means one of the characters are same, keep comparing then
            if sorted_index == indexes and len(set(sorted_index)) != len(sorted_index):
                return "Keep Comparing"
            
            if sorted_index != indexes:
                return False
            if sorted_index == indexes:
                return True
            

        i = 0
        while i < max_loop_length:
            chars = []
            for word in words:
                if i >= len(word):
                    chars.append(" ")
                else:
                    chars.append(word[i])
            
            compare_res = check_lex(chars)
            if compare_res != "Keep Comparing":
                return compare_res

            i += 1
        
        return True
        
        
```
