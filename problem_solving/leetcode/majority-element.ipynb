```python
from typing import List
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        
        elements_count = defaultdict(int)

        for i in nums:
            elements_count[i] += 1
        
        elements_count = [(x, y) for x, y in elements_count.items()]

        elements_count = sorted(elements_count, key=lambda x: x[1], reverse=True)

        return elements_count[0][0]
```
