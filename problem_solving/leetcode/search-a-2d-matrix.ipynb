Link: https://leetcode.com/problems/search-a-2d-matrix/

```python
from typing import List
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        row = len(matrix)

        def binary_search(arr):
            left, right = 0, len(arr)-1

            # print("arr", arr)
            if len(arr) == 1:
                return arr[0] == target

            while left <= right:
                mid = (left+right)//2
                if arr[mid] == target:
                    return True
                elif arr[mid] > target:
                    right = mid-1
                else:
                    left = mid+1
            
            return False


        for i in range(row):
            max_row_element = matrix[i][-1]
            # The element is in ith row
            # print(f"{target} <= {max_row_element}")
            if target <= max_row_element:
                return binary_search(matrix[i])

        return False
```
