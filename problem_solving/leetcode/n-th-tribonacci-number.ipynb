Link: https://leetcode.com/problems/n-th-tribonacci-number/

```python
class Solution:
    def tribonacci(self, n: int) -> int:
        trib = [0,1,1]

        # For first three tribonacci numbers
        if n < len(trib):
            return trib[n]

        for i in range(3, n+1):
            trib.append(trib[i-1] + trib[i-2] + trib[i-3])

        return trib[-1]
```

```python
class Solution:
    def tribonacci(self, n: int) -> int:    
        a,b,c = 0,1,1

        if n == 0:
            return a
        if n in [1, 2]:
            return b

        for i in range(3, n+1):
            a,b,c = b,c,a+b+c
        
        return c
```
