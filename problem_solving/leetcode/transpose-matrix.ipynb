```python
from typing import List
class Solution:
    def transpose(self, matrix: List[List[int]]) -> List[List[int]]:

        m, n = len(matrix), len(matrix[0])
        res = [[0]*m for _ in range(n)]
        
        i, j = 0, 0
        while i < n:
            j = 0
            while j < m:
                res[i][j] = matrix[j][i]
                j += 1
            i += 1
        
        return res
```
