Link: https://leetcode.com/problems/unique-email-addresses/

```python
from typing import List
```

```python
class Solution:
    def numUniqueEmails(self, emails: List[str]) -> int:
        res = []
        for e in emails:
            local_name, domain = e.split("@")
            if "+" in local_name:
                plus_splits = local_name.split("+")
                local_name = plus_splits[0]
            
            if "." in local_name:
                local_name = "".join(local_name.split("."))

            new_email = f"{local_name}@{domain}"
            if new_email not in res:
                res.append(new_email)

        return len(res)
```
