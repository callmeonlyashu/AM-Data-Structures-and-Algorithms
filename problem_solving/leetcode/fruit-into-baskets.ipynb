### Optimized Approach O(n)

```python
from typing import List
from collections import defaultdict
class Solution:
    def totalFruit(self, fruits: List[int]) -> int:
        count = defaultdict(int)
        l = 0

        for r in range(len(fruits)):
            count[fruits[r]] += 1
            # print("count", dict(count))
            if len(count) > 2:
                count[fruits[l]] -= 1
                if count[fruits[l]] == 0:
                    count.pop(fruits[l])
                l += 1

        return len(fruits) - l


        # i, j = 0, 0

        # window_length = 0
        # window = []
        # while j < len(fruits):
        #     window.append(fruits[j])
        #     curr_length = (j-i)+1
        #     if len(set(window)) <= 2:
        #         window_length = max(window_length, len(window))
        #     else:
        #         i += 1
        #         del window[0]
            
        #     j += 1
        
        # return window_length
```

### Unoptimized Approach O(n2)

```python

from typing import List
from collections import defaultdict
class Solution:
    def totalFruit(self, fruits: List[int]) -> int:
        i, j = 0, 0

        window_length = 0
        window = []
        while j < len(fruits):
            window.append(fruits[j])
            curr_length = (j-i)+1
            if len(set(window)) <= 2:
                window_length = max(window_length, len(window))
            else:
                i += 1
                del window[0]
            
            j += 1
        
        return window_length
```
