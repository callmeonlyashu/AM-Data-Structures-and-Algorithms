```python
class Solution:
    def divide(self, dividend: int, divisor: int) -> int:
        MAX_INT = 2**31 - 1
        MIN_INT = -2**31

        # Handle overflow case
        if dividend == MIN_INT and divisor == -1:
            return MAX_INT

        # Determine the sign of the result
        negative = (dividend < 0) != (divisor < 0)

        # Work with absolute values
        dividend, divisor = abs(dividend), abs(divisor)
        quotient = 0

        # Bit manipulation loop
        for i in reversed(range(32)):  # from 31 down to 0
            
            print(f"(divisor << i): {(divisor << i)} and dividend: {dividend}")
            if (divisor << i) <= dividend:
                dividend -= (divisor << i)
                quotient += (1 << i)

        return -quotient if negative else quotient

    
if __name__ == '__main__':
    sol = Solution()
#     result = sol.divide(10, 3)
#     print(result)
#     result = sol.divide(7, -3)
#     print(result)
    
#     result = sol.divide(-2147483648, -1)
    
#     result = sol.divide(2147483647, 2)
    
    result = sol.divide(2121, 2)
    print(result)
```
