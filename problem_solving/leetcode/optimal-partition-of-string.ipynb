Link: https://leetcode.com/problems/optimal-partition-of-string/

```python
class Solution:
    def partitionString(self, s: str) -> int:
        
        sub_str = ""
        res = []
        for i in range(len(s)):
            if s[i] not in sub_str:
                sub_str += s[i]
            else:
                res.append(sub_str)
                sub_str = s[i]
            
        if len(sub_str) > 0:
            res.append(sub_str)
        return len(res)
```
