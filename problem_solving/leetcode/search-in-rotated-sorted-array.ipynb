```python
from typing import List
class Solution:
    def search(self, nums: List[int], target: int) -> int:
#         l = 0
#         r = len(nums)-1
        
#         # If only one element is there in array
#         if r == 0:
#             if nums[r] == target:
#                 return r
#             else:
#                 return -1
        
#         target_index = -1
#         while l < r:
#             mid = (l+r)//2
#             print(f"l: {nums[l]}, mid: {nums[mid]}, r: {nums[r]}")
#             # Check all the three if target is in left right or center
#             if nums[mid] == target:
#                 target_index = mid
#                 break
            
#             if nums[r] == target:
#                 target_index = r
#                 break
            
#             if nums[l] == target:
#                 target_index = l
#                 break
            
#             if nums[l] <= target <= nums[mid]:
#                 r = mid
#             else:
#                 l = mid + 1
#         print(f"RESULT: {target_index}")      
#         return target_index

        l = 0
        r = len(nums)-1
        
        target_index = -1
        
        # If only one element is there in array
        if r == 0:
            if nums[r] == target:
                target_index = r
            else:
                target_index = -1
        
        while l <= r:
            mid = (l+r)//2
            
            print(f"nums[l]: {nums[l]}, nums[mid]: {nums[mid]}, nums[r]: {nums[r]}")
            print(f"l: {l}, mid: {mid}, r: {r}")
        
            if nums[mid] == target:
                target_index = mid
                break
            
            if nums[r] == target:
                target_index = r
                break
            
            if nums[l] == target:
                target_index = l
                break
            
            # Check if mid is in left or right of array
            if nums[l] <= nums[mid]:
                # Compare with the leftmost element to see if we need to look to the left or right,
                if nums[l] < target < nums[mid]:
                    r = mid
                else:
#                     print("here")
                    l = mid + 1
            else:
                if nums[mid] < target < nums[r]:
                    l = mid + 1
                else:
#                     print("here")
                    r = mid
        
        print(f"RESULT: {target_index}")
        return target_index
                
    
s = Solution()
assert s.search([4,5,6,7,0,1,2], 0) == 4
assert s.search([4,5,6,7,0,1,2], 3) == -1
assert s.search([1], 0) == -1
assert s.search([1], 1) == 0
assert s.search([1,3,5], 3) == 1
assert s.search([5,1,3], 3) == 2
assert s.search([5,1,3], 5) == 0
assert s.search([5,3], 3) == 1
assert s.search([4,5,6,7,0,1,2], 5) == 1
assert s.search([4,5,6,7,0,1,2], 4) == 0
assert s.search([4,5,6,7,0,1,2], 6) == 2
assert s.search([5,1,2,3,4], 1) == 1
```
