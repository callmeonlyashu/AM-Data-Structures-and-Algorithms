Link: https://leetcode.com/problems/find-the-town-judge/

```python
from typing import List
class Solution:
    def findJudge(self, n: int, trust: List[List[int]]) -> int:
        if not len(trust):
            if n == 1:
                return 1
            else:
                return -1

        suspects = []
        can_be_judge = []
        for j in trust:
            if j[1] not in can_be_judge:
                suspects.append(j[1])

            can_be_judge.append(j[0])
            if j[0] in suspects:
                new_list = [item for item in suspects if item != j[0]]
                suspects = new_list.copy()
            
        if len(suspects) == (n-1) and len(set(suspects)) == 1:
            return suspects.pop()
        else:
            return -1
```
