```python
class MyStack:
    def __init__(self):
        self.mystack = []

    def __str__(self):
        return f"Stack: {self.mystack}"

    def push(self, value):
        self.mystack.append(value)

    def size(self):
        return len(self.mystack)

    def is_empty(self):
        return not self.size()

    def peek(self):
        popped = None
        if not self.is_empty():
            popped = self.mystack[-1]
        
        return popped

    def pop(self):
        popped = None
        if not self.is_empty():
            popped = self.mystack[-1]
            del self.mystack[-1]
        
        return popped

class MyQueue:

    def __init__(self):
        self.q = MyStack()
        

    def push(self, x: int) -> None:
        stack = MyStack()

        while not self.q.is_empty():
            stack.push(self.q.pop())
        
        stack.push(x)

        while not stack.is_empty():
            self.q.push(stack.pop())


    def pop(self) -> int:
        return self.q.pop()
        

    def peek(self) -> int:
        return self.q.peek()

    def empty(self) -> bool:
        return self.q.is_empty()


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

```python
class MyStack:
    def __init__(self):
        self.mystack = []

    def __str__(self):
        return f"Stack: {self.mystack}"

    def push(self, value):
        self.mystack.append(value)

    def size(self):
        return len(self.mystack)

    def is_empty(self):
        return not self.size()

    def peek(self):
        popped = None
        if not self.is_empty():
            popped = self.mystack[-1]
        
        return popped

    def pop(self):
        popped = None
        if not self.is_empty():
            popped = self.mystack[-1]
            del self.mystack[-1]
        
        return popped

class MyQueue:

    def __init__(self):
        self.in_stack = MyStack()   # for incoming elements
        self.out_stack = MyStack()  # for outgoing elements

    def push(self, x: int) -> None:
        self.in_stack.push(x)

    def pop(self) -> int:
        if self.out_stack.is_empty():
            while not self.in_stack.is_empty():
                self.out_stack.push(self.in_stack.pop())
        return self.out_stack.pop()

    def peek(self) -> int:
        if self.out_stack.is_empty():
            while not self.in_stack.is_empty():
                self.out_stack.push(self.in_stack.pop())
        return self.out_stack.peek()

    def empty(self) -> bool:
        return self.in_stack.is_empty() and self.out_stack.is_empty()


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
