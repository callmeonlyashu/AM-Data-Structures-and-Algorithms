```python
"""
You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

Merge nums1 and nums2 into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

 

Example 1:

Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]
Explanation: The arrays we are merging are [1,2,3] and [2,5,6].
The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.


Example 2:

Input: nums1 = [1], m = 1, nums2 = [], n = 0
Output: [1]
Explanation: The arrays we are merging are [1] and [].
The result of the merge is [1].


Example 3:

Input: nums1 = [0], m = 0, nums2 = [1], n = 1
Output: [1]
Explanation: The arrays we are merging are [] and [1].
The result of the merge is [1].
Note that because m = 0, there are no elements in nums1. The 0 is only there to ensure the merge result can fit in nums1.

"""
```

```python
"""
âœ… Optimization Hint (No Code, Just Concept)

ðŸ§  Step-by-Step Algorithm
	1.	Initialize pointers:
	â€¢	p1 = m - 1 â†’ points to the last valid element in nums1
	â€¢	p2 = n - 1 â†’ points to the last element in nums2
	â€¢	p = m + n - 1 â†’ points to the last position in nums1
	2.	Loop:
	â€¢	While both p1 >= 0 and p2 >= 0:
	â€¢	If nums1[p1] > nums2[p2]:
	â€¢	Set nums1[p] = nums1[p1]
	â€¢	Decrement p1
	â€¢	Else:
	â€¢	Set nums1[p] = nums2[p2]
	â€¢	Decrement p2
	â€¢	Decrement p
	3.	Leftover elements in nums2:
	â€¢	After the loop, if any elements remain in nums2, copy them:

"""

from typing import List
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        p = m+n-1
        p1 = m-1
        p2 = n-1
        while p1 >= 0 and p2>=0:
            if nums1[p1] > nums2[p2]:
                nums1[p] = nums1[p1]
                p1 -= 1
            else:
                nums1[p] = nums2[p2]
                p2 -= 1
            p -= 1
        
        while p2 >= 0:
            nums1[p] = nums2[p2]
            p2 -= 1
            p -= 1
                
s = Solution()
nums1 = [1,2,3,0,0,0]
s.merge(nums1, 3, [2,5,6], 3)
assert nums1 == [1,2,2,3,5,6]

nums1 = [1]
s.merge(nums1, 1, [], 0)
assert nums1 == [1]

nums1 = [0]
s.merge(nums1, 0, [1], 1)
assert nums1 == [1]

nums1 = [4,5,6,0,0,0]
s.merge(nums1, 3, [1,2,3], 3)
assert nums1 == [1,2,3,4,5,6]

nums1 = [2,0]
s.merge(nums1, 1, [1], 1)
assert nums1 == [1,2]

nums1 = [0,0,0]
s.merge(nums1, 0, [2,5,7], 3)
assert nums1 == [2,5,7]

nums1 = [-1,0,0,3,3,3,0,0,0]
s.merge(nums1, 6, [1,2,2], 3)
assert nums1 == [-1,0,0,1,2,2,3,3,3]

nums1 = [0,0,3,0,0,0,0,0,0]
s.merge(nums1, 3, [-1,1,1,1,2,3], 6)
assert nums1 == [-1,0,0,1,1,1,2,3,3]


nums1 = [-1,3,0,0,0,0,0]
s.merge(nums1, 2, [0,0,1,2,3], 5)
assert nums1 == [-1,0,0,1,2,3,3]

nums1 = [-1,0,0,0,3,0,0,0,0,0,0]
s.merge(nums1, 5, [-1,-1,0,0,1,2], 6)
assert nums1 == [-1,-1,-1,0,0,0,0,0,1,2,3]

nums1 = [1,2,0,0,0]
s.merge(nums1, 2, [0,3,4], 3)
assert nums1 == [0,1,2,3,4]

print("all test passed")
        
```
