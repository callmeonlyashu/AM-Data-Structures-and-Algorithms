Link: https://leetcode.com/problems/isomorphic-strings/

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        # Check if length of strings are matching
        if len(s) != len(t):
            return False

        mapper = {}
        rev_mapper = {}

        i, j = 0, 0
        while i < len(s) and j < len(t):
            if s[i] not in mapper:
                mapper[s[i]] = t[j]
            if t[j] not in rev_mapper:
                rev_mapper[t[j]] = s[i]

            if mapper[s[i]] != t[j]:
                return False
            if rev_mapper[t[j]] != s[i]:
                return False
            
            i += 1
            j += 1
        
        return True
```
