Link: https://leetcode.com/problems/text-justification/

```python
from typing import List
class Solution:
    def fullJustify(self, words: List[str], maxWidth: int) -> List[str]:
        res = []
        line = []
        curr_len = 0
        for word in words:
            if curr_len + len(word) + len(line) > maxWidth:
                res.append(line)
                line = []
                curr_len = 0
            line.append(word)
            curr_len += len(word)

        if line:
            res.append(line)

        final_res = []
        for i, ln in enumerate(res):
            words_length = sum(len(w) for w in ln)
            space_needed = maxWidth - words_length
            no_of_spaces = len(ln) - 1

            if i == len(res) - 1:
                s = " ".join(ln)
                s += " " * (maxWidth - len(s))
                final_res.append(s)
                continue

            if no_of_spaces == 0:
                final_res.append(ln[0] + " " * space_needed)
                continue

            base = space_needed // no_of_spaces
            extra = space_needed % no_of_spaces

            new_str = []
            for idx, w in enumerate(ln):
                new_str.append(w)
                if idx < no_of_spaces:
                    spaces_to_add = base + (1 if idx < extra else 0)
                    new_str.append(" " * spaces_to_add)
            final_res.append("".join(new_str))

        return final_res
```
