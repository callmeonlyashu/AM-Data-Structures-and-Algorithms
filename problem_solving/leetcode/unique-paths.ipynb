Link: https://leetcode.com/problems/unique-paths/

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        memo = {}
        def dfs(i,j):
            if i == m-1 and j == n-1:
                return 1

            if i >= m or j >= n:
                return 0

            memo_index = (i,j)
            if memo_index in memo:
                return memo[memo_index]

            memo[memo_index] = dfs(i+1, j) + dfs(i, j+1)
            return memo[memo_index]

        paths = dfs(0,0)
        return paths
```
