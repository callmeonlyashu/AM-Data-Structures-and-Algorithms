Link: https://leetcode.com/problems/clear-digits/

```python

class Solution:
    def clearDigits(self, s: str) -> str:
        stack = []

        for d in s:
            if d.isdigit() and len(stack):
                stack.pop()
            else:
                stack.append(d)

        return "".join(stack)
```

```python
s = Solution()
res = s.clearDigits("cb34asads")

print(res)
```
