Link: https://leetcode.com/problems/removing-stars-from-a-string

```python
class Solution:
    def removeStars(self, s: str) -> str:
        stack = []
        for i in s:
            if i == "*" and len(stack):
                del stack[-1]
            else:
                stack.append(i)
        
        res = "".join(stack)
        return res
```
