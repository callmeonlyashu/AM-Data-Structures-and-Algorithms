Link: https://leetcode.com/problems/maximum-number-of-vowels-in-a-substring-of-given-length/

```python
class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        vowels = {'a', 'e', 'i', 'o', 'u'}

        def count_vowels(st):
            count = 0
            for w in st:
                if w in vowels:
                    count += 1
            return count

        first_window = s[:k]
        res = count_vowels(first_window)
        max_res = res
        i = 0
        for j in range(k, len(s)):
            if s[j] in vowels:
                res += 1
            if s[i] in vowels:
                res -= 1
            
            i += 1
            max_res = max(res, max_res)
        
        return max_res

```
