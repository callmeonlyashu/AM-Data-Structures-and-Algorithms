Link: https://leetcode.com/problems/can-place-flowers/

```python
from typing import List
class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        curr = 0

        res = 0
        while curr < len(flowerbed):

            if curr != 0:
                prev = curr-1
            else:
                prev = curr

            if curr < len(flowerbed)-1:
                next = curr+1
            else:
                next = curr

            if flowerbed[prev] == flowerbed[curr] == flowerbed[next] == 0:
                res += 1
                flowerbed[curr] = 1

            curr += 1

        return res >= n
```

## Better Version

```python
class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:

        f = [0] + flowerbed + [0]

        for i in range(1, len(f) - 1):
            if f[i - 1] == 0 and f[i] == 0 and f[i + 1] == 0:
                f[i] = 1
                n -= 1

        return n <= 0
```
