```python
from typing import List
class Solution:
    def decrypt(self, code: List[int], k: int) -> List[int]:
        def get_numbers(idx):
            nums = 0
            for _ in range(abs(k)):
                if k > 0:
                    if idx+1 >= len(code):
                        idx = 0
                    else:
                        idx = idx+1

                elif k < 0:
                    idx = idx-1
                
                nums += code[idx]

            return nums

        res = []
        for i in range(len(code)):
            if k != 0:
                res.append(get_numbers(i))
            else:
                res.append(0) 
            
        return res
```
