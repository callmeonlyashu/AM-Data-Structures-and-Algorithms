```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def sumNumbers(self, root: Optional[TreeNode]) -> int:
        root_val = root.val

        res = []
        def dfs(root, numbers=[]):
            if not root:
                return

            if root and root.left is None and root.right is None:
                numbers.append(str(root.val))
                if numbers:
                    num = int("".join(numbers.copy()))
                    numbers.pop()
                    res.append(num)

            numbers.append(str(root.val))
            dfs(root.left)
            dfs(root.right)            
            numbers.pop()

        dfs(root)
        return sum(res)

```
