```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
    
class Solution:
    def insertGreatestCommonDivisors(self, head: Optional[ListNode]) -> Optional[ListNode]:
        curr = head
        left = head
        right = head

        def gca(num1, num2):
            minimum = min(num1, num2)
            max_div = 1
            for i in range(2, minimum+1):
                if num1%i == 0 and num2%i == 0:
                    max_div = max(max_div, i)
            
            return max_div

        while right.next:
            right = right.next
            gcd = 1
            if right:
                gcd = gca(left.val, right.val)

            left.next = ListNode(gcd, left.next)

            left = left.next.next
        
        return head
```
