Link: https://leetcode.com/problems/maximum-nesting-depth-of-the-parentheses/

```python
class Solution:
    def maxDepth(self, s: str) -> int:

        depth = 0
        max_depth = 0
        for ch in s:
            if ch == ")" and depth:
                depth -= 1
            elif ch == "(":
                depth += 1
        
            max_depth = max(max_depth, depth)
        return max_depth
```
