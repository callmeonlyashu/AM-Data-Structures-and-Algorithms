Link: https://leetcode.com/problems/lemonade-change/

```python
from typing import List
class Solution:
    def lemonadeChange(self, bills: List[int]) -> bool:
        denominations = {20: 0, 10: 0, 5: 0}

        for bill in bills:
            denominations[bill] += 1
            change = bill - 5
            while change > 0:
                if change >= 20 and denominations[20] > 0:
                    change -= 20
                    denominations[20] -= 1
                if change >= 10 and denominations[10] > 0:
                    change -= 10
                    denominations[10] -= 1
                if  change >= 5 and denominations[5] > 0:
                    change -= 5
                    denominations[5] -= 1
                else:
                    return False

        return True
            
```
