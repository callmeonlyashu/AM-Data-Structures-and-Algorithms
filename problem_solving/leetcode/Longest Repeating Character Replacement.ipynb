```python
from collections import defaultdict
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        s_length = len(s)
        if s_length < k:
            return s_length
        
        i = 0
        j = 0
        max_length = 0
        freq = defaultdict(int)
        max_count = 0

        # Create window using i and j
        while j < s_length:
            freq[s[j]] += 1
            max_count = max(max_count, freq[s[j]])

            # Calculate how many characters need replacement
            num_replacements = (j - i + 1) - max_count

            if num_replacements <= k:
                max_length = max(max_length, j - i + 1)
                j += 1
            else:
                freq[s[i]] -= 1
                i += 1
                j += 1

        return max_length

s = Solution()
assert s.characterReplacement("ABAB", 2) == 4
# assert s.characterReplacement("AABABBA", 1) == 4
# assert s.characterReplacement("ABCDE", 2) == 3

# long_string = "BBAABABABAAAAABBABBABAAAABABAABAABABABAABABBAAABBAABBBBABAABBABBABBABBABABBAABBABAABBBBBABAAABBBABABABBBAAABAABBABABBBAAABBBBBBAAAAAABBBBABBBABBABAABABABBBAAABBAABBABABBAABABBAABABAABBABABABABBAAAABABBBABABAABABBAAABAABABBBAAAAABAAABAAAABABBAAAAAAAAAABBBBAABAABBAAAAAAABBBABABABBBABAAAABAABBBBABAABAAABBAAABAABAAAAABBAAAAABAABABABBBABAABABBBABAABBBAABABAABBAAAAAABAAAAABBBAAAAAAAAAAAAABABAABBBABBAABABBABBAAABBBBBAABBABAAABBBAABABAABAAAAAAABABAAABABABABBABABBBBBBAABBBABBBAABAABBBBBABBBBAAABABAABBAABAABABAAABAAAABBBBBABBAABBABBAAAABBABABBABBBBABBABBBAAABAABABBBABBBAAABAAABBBBBABAABABABAAAAABBBBAAABABBAAABBABAABBABBABABAAAAAABAAAAABABBAABAABAAAABAABABBBBAAAAAABAABAABAAAABAAAAABABAABBBAAAABBABABBBABBBABBBAABBABABAABABAABBBAABBABBABBAAAABAABABBBABBBBABABABAABABBABAABABBAAAAAABBABAABABABAAABBBBABBAAAABABBABBABAAABBABBABBAAABABBAAABBABABBBABBAABBAABABBBBBBBAAAAABAABABAABAAABABBABABBAABBAABAAABBABBBAABBABBABBBBABBAAAAABBBABBAAAABBABBBABAAAABBAAAAAAABBAAABBABABABBAAAABBBAAABAABABBABAAAAAAABAAAABAAAABABABBBABBBBBAABBABBAABAABAABAABAAABAABAAAAABBAAABBBBABAABAAABAAABBABBBBBBBABBBBABAABAABBAAABABBAAABBBABBBBAAAAAAAAABAABAAAAABAAABBBAABABBBBBAAABBAABBBABAABAAAABABBABBBBABABBABBBBBABBBBAAABBABABBBABBBABBBAABABABAABABBABAABABBBABBABBBBAABABABABABBABAABBBBBAAABBBBAABABBAABAABBAAABBBBBBBBBBABBAAABBBABABBBABAAAABBBABABABBABAAAAAABBABBABBBBBBBBAAABAAAABBBABBBBABAABABAAAABABABBAAABABAAAABBBAABABBAABABABAABBABABAABBAABBBABABABAAAAAAAABAAAAABBAABBBAAAABBBAABBAAAABBAAAABBABBAABAAABBBAAABBABABBABBBBBABAAAABBBABABABABBBBBBBAAAABAABBBBABAABBBBBAABAAABBABABBBABBAABABBAAAAAAAABAABABBBABAAABBBBAAAAAABBABABBAAAAAAABAABAABABAAABBBAABBAAAAAABBAAABABBBBAAAABABBABABABABBAAABAAAABBBBBAABABBBABBAABBAAABBAAABAABBAAAABBABBABBBBAAABABABBABAABBAAABBAAAAAABABBABBBABBAABBBBAAABBAAABAABAABBBBBAAAABAAAABABAAAAAABAAAAAABBBBBAAAABABAAAABABBBAAABAAABAABAABABBAAABBBBABAABBAAABBAAAAAABAABAABBABABABBABABAAABABAAAAAAAAAABABAAABAAAAABBAABAAABAAABAABABBBABBAAABBABBBABBBBAAAAAAAABAABBBABBBBAABABAABAAABABBBABBAAABBBBBBBBABABBBBAAABABBABABABABABABABBAAAAABBAAABAABBBABAABABBBAAABBBABABBBBBBBBAAABAABABBABABBBAABBABBBBBBBAAABAAABABAABABBBABAAABAAABAABBBAABBBBBBABBAABAABABBABAABAABBBBBAABBBAABAAABAABABBAAABBBBBABBBBBBBABBBAABBAABBBBBAABBBAAABABBABBABBBABAABBBBAAAAABBAABBBBBABAAABABBABAAABAABBBABAABAABABABBAAAABAAABABBABBBBABBABBBBAABABABAABAABBBABBBBBABBAABAABBBBAABBBBAAAAABAAAABAABBBABBABBBBBBAABBAAABBBBAABBBABABAAAAABABABBABABABBBAABBABBABBAAAAABBABABAABABBBBAABBBBAABABABAABBABABBAABBBBBBBBBBAABBBABAAABBBBBBABBBBABBBBBAAABABBBABBBBABBABBBAAAAABBABABBBBAABBAAAABABAABAAABBABABBBBAAABAABBBBBAAABBBAAABBBBBBBBBAAABABBBBBBBBABABBBABBBBAAAABAAABBBAABABAAAABBAAABBABBBBBBAAABBAABAAAAABABAAAAAABABBAAAABBABAABAABBBBABAAAABBABBBABBBABAAABABAABAAAAAAAABBBABAAABBAAABBBABABBAABBAABAABBBAABBAAABBABAAABAAAABBBBBBBAABABAAABBBABBBBABBABBABBBABBABAABABABAAABBAAABABBBBBAAAABBBAABAABAAABAAABAAABBBBBBBAABBBABABBABAABAAABBAAAAAAAABABABBBABBABBBBBABABBBABBBBAABBBBABAABAABBAABBABAAAABBBABBBABAAAABBBABAABBBABABBBBAAAABBBAABBBABABBABBAAAAABABABBBAAABAAAAAABAABBABBAAABBABBABBBAABABAABABBAAABBBBAABABBABBAAAABABBAAAABBBAABAAAABAAAAAAAABBABBBABABBAAAAAABBBABAAABBABAABAAABAAAABBABBBBABAABBBABAABAABBBAABAAAAABBABABABBBBBAAABBBABBAAAABBAABBAAAABBBAABAAAAABBABABABABABABBBBBBAAAAABBBAAAABAABBAAAAAAAAAABBBAABAABBBBABBABAAABABABAABAAABBAAAABABABABAAABABAABBABAABAABBABBAABAABABBAAABABBBBBABBBBAABAAABBAABBBAABBAAAAABBBAAAAAAABBABBBBBBAABBABABAAAAABBBABABBBAABABAAAABABBABAAAAABBABBBAABBBAAAAAAAABBBBABBABABAABABAABBBAABAABAAABAAABBAABBABABABBBBBAABAABBAAABABBAAAABBABAAABABBBBABBAABBAABBBBABBBABAAABABAABBBBBBAAAABAAABABABAAAABABAABAABABAABBBABAABABBAAAAAAABABAAABAAABABABAAABABBBABBABBBBBBBABBBAABBBABBBBABAABBBABABBBBBBBBABABBAAABABBAAAABBBABABBABBBBAAABABBBAAAAAAABAABABBBBBBBBBBAABABBABBBABBAAAABBAABBBBAAAAABAABAAABAABBABAABBABAAABBAAAABBBBBBBABAABABBBAABBBABBBABAABAAABBBBABBBABAAABAAAAAABBAAAAAAAABAABBAAAABABAAABAAABBABABBAAABABBBABABBBAABAAABABBAAAABBAABABBAABBBAABBABABAABAABBAAAAABAABBBAAAABBBBBBBBABAABAAAAABBAABBAABBAABABBABBAABBBBAABBBABBBAAABABABABBAAABAAAABBBBBBABBAABBABABBBABAAABAABAABABBABBBBBAAABABABABBAABABABBAAAAAAAABABBBABABABBAAAAAABBBBBBAABABBABABAAAAAAABABBBBBAAAAABBBBABABAABBBBABABBAABABBBBBABAABAABBAAABBAAAAAABBABBABBAABABBBAABBABABAABAAABAABABBABBABAAABBBBABAAAAABBBAABAAAABBBBBBAABAAAAABAAABABBBAAAAAABAABAAABAAABAABBBABABBABBABBBAABBBBAABAAABBBBAABBBAABABBBBBBAAAABBBABABBAAAAABBABBBBBBABBBBABAAABBBBBBBABAABABBBBABABABAAAAAABABBBAAAABAABAABBABAABABBBBBABBAABABAAABABBBBBBAABBABABBBABAABBAAABAABAAAABABAAABABAAABABABABABBABBAABBAABABBABAAABBBABAABBBBAAABBAAAAAABAAAABAAAABBAAAAAABBAAAAABBBBBBBBABBAAABAABBBBAABBAABBBABABABBBAAAAAAAABBBBBABBABBAABBBABABBBBAAAAABBABBAABBABABAABABABAAABAAABBABAABABBABAABBABBAABBAABBABABBBBBABBBABABBABAAABBBAABBBBABAABAAABBBBBBBAAAABAAABABBBBABABABAAABABABBAABBAAABABBBBABABAAAAABABAAABABBBBBAAABBBAABAABAABBBBABBBBBABABAABABABBBBAAAABBAABABABBBBBBBBBBAAAABAAAAABBBABBAABAAABAABAABAABABBBAAABBBABBAABAABAABBAAABABAAABAAAABAAAABABABABBAABABBBABBBBAAAABABABAABBBAAABBBABABBABBAAAABABABABBBAAABABBABAABBAAAABABABAAABAABBABBAABABBBBABBAABABBBABABBBABAABABBAABABABAABBABBBBABBAAAABAABABAABABABBBABAAABAAAAABBABABAABBBBBBABBABBABBBABAAAABAAAABABBBBAABABBAAABBAABBBBAAAABABBABABABAAABAAABBBBBABAAAABAAAABBBBBAABABAAAABBAAAABAABAAABAABAABABBBBBAAABAAAAABAAAABBBAABAABAABBBBABBABBAABBBAAAABBBBAABBAABABAABBBAAAAABABBBBABABBBABBABAAABAABBBABAAAABABABBBABAAABBABAAABABABABBABAABAAAABABBAAAAAABAAAABBBBABBABBBBBABABAAABBAABAAABBAABBAAABBBBBBBBBABBAABBBBAABABBAAABBBAAAABABAAAABAABAAAABBBBABBBBABBAABAAABABBBAABABBBBAAAAAABAAABAABAAAAABBAABABABABBAABBBABABBABBABBABAAABBAAABABBBBAAAABBAAABABBBABAABBABAAABABABBABBBBBAABBABBAABBBAAAABABAAAAAABABAAAAABAAABABAAABABABAAABBABBAAABBBAABBABBABABAABBBABABBABABBBAABAAAAABABBAABAAABAAABBBBAABBBAABBAABABABBABBABAAABAABBABBAABBABBAAAABBBBAABAAABBBAAAABAAAABABBBAAABAAAAABBAABABBABABAAAAAAAAABABAAAAABABABBBBABBBBAAAAABBAAABABAABAABABAAAAAAAAAAAAABAABAAABBABAAAAAABAABABBABBBAAABABAAABABABBABBBAAABBBABAAAAAAAABAAABABBAABBBBABBAAABAAABABBBAAAAAABABAABABBBAABBBABABBBBBBAABABABABABBAAABAABBABABAAAABABAABABBAAABAAAAAABBABBBAAABABAABBAAAAAAAAAABBAABBBABAABBABBAAAAABABAAABABAABAABBABABBABBBBBAABBAAAABABBBABBABABBABABBAAAABBAAAABABABBBBBBABBBBBBBBAABABBABABABABAAAAAAAAABAABAABBBBBABABAAABABAAAAAABABAAABAABBAAABAABBBAABBAABAABABBBBBBBABBBBABABAABAABABABAAABBBAAABABABAAABAABBBABABABBABABABAABBBBAABBBBBBBABBBABAAABABABBBABBBAABBBAABBAAABABBBBABBABBABABBBABAABBBABBABAABABBBBAABABBBABBAABAABABAABBBBAAAAAABBBABBBBBBAAAABBABABBBBBBABABAABBBBABABABAABBAABBAAABABBAABABBABBABBABABABAAABAAAAAAABABBBBABBBABAAABABBAAAAABAABBBAAAAAAABBAAAAAAAABBAABAAAAABBBABAAAABBBAAAABBBBABBAABAAABBBBAABBABABABABAAABBABBAAAAAABBBABBABAAABABBBABAABBBBBABBBABAAABBBBABBBBAABBABBABBBBBAABABBBBBABBABBBABBBBAAABABBAABABBAABBBBBBABABBABABBBBAABABABABBBAAAAABABABBABBBAAAAAAABAABABABBAABAAAAABAABAABBAABBBAABBBBBBBBABBBBBBBAAABABABBBBBABABAAAAABBABABABABAABABABBBBBABAABBABABAAABABBBABAAAAABBAAAAABAAAAABAABBAABBBABBAABAABAAABABBABAABBBAABAABBAABABABBBBAABBBABBAAAABAABABBABBBAAABBBAABAAAAAABABAABAAABBABBABBAAAAAAABBBAAABBBBBAABBAAABAABBAABBAAAAABBABBBBBBBBBBBBABAABBABAAAABBAAABBAABABBBABABBAAABBBAAAAAAABABBAABABBAABABBBBBAABABBBAAAABAAABAAABABAAAAABABAAABAAAAABBABABBAAAAAABAAAABBABABAAABABAAAAABAABBBABBBBBABBBABAAABABAABAABBAAAAAAABBBBABAABABABBABABABBBBABBABBABBABBBBBBABBBAABBABBBAABAABBAABBABABAAAAAAAABAABBBBBBBABBBBAABAAABAAAABABABABAABBBAABABABBAABAABBAAABABABABBAABAAAABBBBAAABBAABAABABABBABBBBBABABABBAABBBAABBAAAABAABABBAABBBBBBAAAABABBAAAABABBABABAABBBBBBBBBABABAAAAAAABBABAAAAAAAABABBBBAABBAABABAABBBBBBAABABAAAABBAABAABBAABABABBABBAABBAAABBABAABBABAABBAAABBABBBAABABBABBBBBABBABBAAABBBBBABAAAAAABABAABBABBABAABBAAABABAABAAAABAAABBAAABABABABAABBBABBBAAABBBABAABABAAAAAABAABBABAABABBBABABBBBAABBBAABABAABABAABBAABBAABBABABBABAABAABAABBBBABAABBAAABBBBABAAAABBAAAAAAABABABBABABABBBBABABABABBBABAABBABBAABABBABAABBBABAABABAAAABAABBAABBBAABBAAAAAABAAABAAAABBAABBBBBAABABAABABABAAABABABABABABABBAABBBAAABABBABBAABAABABAABAABBBBBBABBAAABBABBBBAAABABBABAABABBBBBBBBBAAABAAAABAAABAABBAAAAAABAAAABBABABBBBAABABAABAABBABBBBBAABABAABAABBBBABBAABBBBAAABAABBBBBAAABAABBABBBBABABBBAAAABBBABBBBBBAABBABAABBBBBAABABAAAAAAABBABABBAABABAABBBAABBAABABAAABABAABABAAAABBBABBBBBABAABAAAABABBBBABBBBAAAAABABBBAABBBABBBAAAAAABABBBBABABBAAABBBBBABAAAABBBABBBBABAABBBBBABBAAABABAAAABABABAABAABBAAABABBAABBBBAAABAAABBBAABBABBBAABBAAABBBBBABBBAAAAAABABBBBABBBBABBBAABBABAABABABABAABAAABBBBBAAAAABABAABAABAAAABAABBBAABAAABBABBABAAAAAAABBBBBBAAAAAAAABBABBBBBBABBABBBBABBBBABABABAABABBABBAAAABBBBBBBABBBAAABBABBBABBBABBAAABBABBAABAAABAAABBBBABAABAABBABABBABAABABABABBBBBAABBBBBAABAAAAAABBABBABBAAAAABBBAABABAABBABAABAABBAABAABAAAABBABABABBBBABBAABBAAABBBAAAAABABAABBBBAABBBBBAABBBBABAAAABAABBAABBABBBAABAABAABAAAABAAABABBABBBAAAAABAABBAABBBBBAAAAAABBAABAAAAABABBBBBAABAABAABAABBAABABBABAABABABABAAAAAAAABBAABAABBABAAAAABABAAAABBBBBAABAAABAAAAAAABBBAABBBABBAAABAAAABABABBAAABAAAAABABABBBBAABAAAABBBAABBABBBABBABABABBAABBAAAAABBBBBABBBBAABBBBBAAAABAABBABABABBBBABBBABBBABBBAABBBAABBBBAABAAAAAAAAAABABBBBABAABBBBAAAAABABAABBBABBAAAAAAABAAABBABABBAAAABAAAAAAABBBBAABBAAABBABAAAAABBBAAAABBBABBBBAABBBBABAAABABBAABABBABBBAAAABAABABBAAAAAABABBBABAAAAABBBBABAAABAAAABBAAAABABABBBABBABAABABBABBBABAAABAABABABABAABBABBBABAABBAABAAAAAABBBBABABABABBBAAABBABABBABBAABBBAABABBBBAAAABABAAAAAABBABBBAAABBABABBABABBBAABAABBBAABABBBBABBBAAABABBBBAABAAABAABBBBABAAABAABABAABABBAAABABBBBBBBBABBBABAABBAAAABBABAABBAAABBBBABBBBAAABBABBBAABAAABAAAAAAAABBAAABAABBBAAABABAAAAABAABBABBAABABBBAABABABBAABAAABABAABABBBBAABAABBBBBABBBBAABABABABBBBABBBBABAABABBABAABBABBABAAABBBABAAABAAAAAABAABABABAAABBBABABAABBBBBABABBAAABABAAABAAABABBABBBBABBBBABAAAAAABBBBAAAAABAAAABAABBABBABBBAABBAABABABAABAAABAABBABAABABBABBBABBAAABBAABBABAABABABBBBAABABBABBAAABABABAAABAAABBABABAAABBAAAAAABBBABABBABAABABABBABBABBAABBBBAAABAAABAABAABABBAABBBAAAAABBBBABABABABBBBBABABBBBAAABABABABBBBABABABABBBBBAABBABAAAABBAABAABAAAABBBBBBAABAAAABBAAABBAABBBAAAABAABAABABABBBBABABAABAABBABAAABBBBABBBBABBBABBABABAAAABBAAAABABABBBBAAAAABABABBBAAB"
s.characterReplacement(long_string, 7886)


print("All Test cases passed")
```
