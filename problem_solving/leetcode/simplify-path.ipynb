Link: https://leetcode.com/problems/simplify-path/

```python
class Solution:
    def simplifyPath(self, path: str) -> str:
        p = path.split("/")
        
        stack = ["/"]
        for folders in p:
            last_folder = stack[-1]
            if folders == '.':
                continue

            if last_folder != "/":
                stack.append("/")

            if folders not in ['..', '']:
                stack.append(folders)

            elif folders == '..' and len(stack) > 1 and len(stack):
                # Two pop to delete / and folder both
                stack.pop()
                stack.pop()


        if stack[-1] == "/" and len(stack) > 1:
            del stack[-1]

        res = ''.join(stack)
        return res
```
