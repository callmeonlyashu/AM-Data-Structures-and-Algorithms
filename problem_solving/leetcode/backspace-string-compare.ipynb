Link: https://leetcode.com/problems/backspace-string-compare/

```python
class Solution:
    def backspaceCompare(self, s: str, t: str) -> bool:
        updated_s = []
        updated_t = []

        for ch in s:
            if ch == '#':
                if updated_s:
                    updated_s.pop()
            else:
                updated_s.append(ch)

        for ch2 in t:
            if ch2 == '#':
                if updated_t:
                    updated_t.pop()
            else:
                updated_t.append(ch2)

        return updated_s == updated_t
```
