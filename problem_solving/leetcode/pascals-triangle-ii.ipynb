```python
from typing import List

class Solution:
    def getRow(self, rowIndex: int) -> List[int]:
        row = [1]
        for i in range(1, rowIndex+1):
            curr_row = [0] + row + [0]
            l, r = 0, 1
            new_row = []
            while r < len(curr_row):
                new_row.append(curr_row[l] + curr_row[r])
                l += 1
                r += 1
            row = new_row.copy()

        return row
```
