```python
class Solution:
    def numberOfMatches(self, n: int) -> int:
    
        teams = n
        total_matches = 0
        while teams >= 2:
            if teams % 2 == 0:
                matches = int(teams/2)
                teams = int(teams/2)
            else:
                matches = int((teams-1)/2)
                teams = int((teams-1)/2) + 1

            total_matches += matches

        return total_matches
```
