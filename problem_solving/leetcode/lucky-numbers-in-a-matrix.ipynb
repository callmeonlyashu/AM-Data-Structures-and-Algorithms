```python
from typing import List
class Solution:
    def luckyNumbers(self, matrix: List[List[int]]) -> List[int]:
        rows, columns = len(matrix), len(matrix[0])
        
        def get_rows_cols(i, j):

            row_elements = []
            col_elements = []
            for k in range(rows):
                col_elements.append(matrix[k][j])

            for l in range(columns):
                row_elements.append(matrix[i][l])
            
            return row_elements, col_elements

        for i in range(rows):
            for j in range(columns):
                r, c = get_rows_cols(i, j)
                if min(r) == max(c):
                    return [matrix[i][j]]

        return []
```
