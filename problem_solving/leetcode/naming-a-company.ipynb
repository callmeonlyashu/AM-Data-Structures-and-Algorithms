```python
class Solution:
    def distinctNames(self, ideas: List[str]) -> int:
        
        groups = collections.defaultdict(set)
        for w in ideas:
            groups[w[0]].add(w[1:])

        res = 0
        for char1 in groups:
            for char2 in groups:
                if char1 == char2:
                    continue

                intersect = sum(1 for w in groups[char1] if w in groups[char2])
                distinct1 = len(groups[char1]) - intersect
                distinct2 = len(groups[char2]) - intersect
                res += distinct1 * distinct2

        return res

```
